```python
import pandas as pd
```

```python
fp = 'data/029740.txt'
df = pd.read_csv(fp, delim_whitespace=True, usecols=['YR--MODAHRMN', 'TEMP'], na_values=['*', '**', '***', '****', '*****', '******'], parse_dates=True, index_col='YR--MODAHRMN')
```

```python
df['TEMP_C'] = (df['TEMP'] - 32) / 1.8
```

```python
df['Month'] = df.index.month.values
df['Year'] = df.index.year.values
```

```python
grouped = df.groupby(["Year", "Month"])
```

```python
years = df.index.year.unique().values
```

```python
years
```

```python
years_ok = []
for i in range(len(years)):
    if years[i]-1 in years:
        years_ok.append(years[i])
```

```python
import numpy as np

winter_means = np.zeros(len(years_ok))
winter_std = np.zeros(len(years_ok))
summer_means = np.zeros(len(years_ok))
summer_std = np.zeros(len(years_ok))

    
```

```python
df.loc[(df.index >= '19521201') & (df.index < '19530301')]
```

```python
for i in range(len(years_ok)):
    winter_min = str(years_ok[i]-1)+'1201'
    winter_max = str(years_ok[i])+'0301'
    summer_min = str(years_ok[i])+'0601'
    summer_max = str(years_ok[i])+'0901'
    winter_means[i] = df['TEMP_C'].loc[(df.index >= winter_min) & (df.index < winter_max)].mean()
    winter_std[i] = df['TEMP_C'].loc[(df.index >= winter_min) & (df.index < winter_max)].std()
    summer_means[i] = df['TEMP_C'].loc[(df.index >= summer_min) & (df.index < summer_max)].mean()
    summer_std[i] = df['TEMP_C'].loc[(df.index >= summer_min) & (df.index < summer_max)].std()
```

```python
winter_means
```

```python
winter_std
```

```python
summer_means
```

```python
summer_std
```

```python
df2 = pd.DataFrame(data={'Year': years_ok, 'Winter mean': winter_means, 'Winter std': winter_std,
                                         'Summer mean': summer_means, 'Summer std': summer_std})
```

```python
df2.to_csv('seasonal_temps2.csv', index=False, float_format='%.6f')
```
